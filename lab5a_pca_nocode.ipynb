```python
model_lbls = ['dt' # decision tree
             #,'nb' # gaussian naive bayes
             ,'rf'   # random forest
             #,'lp'   # linear perceptron
             #,'svc'  # support vector
             #,'knn'  # k nearest neighbours
             #,'adb'  # adaboost
            ]

models = {
    'dt': {'name': 'Decision Tree       ',
           'estimator': DecisionTreeClassifier(random_state=random_state),
           'param': [{'max_depth': [*range(1,20)],'class_weight':[None,'balanced']}],
          },
    'nb': {'name': 'Gaussian Naive Bayes',
           'estimator': GaussianNB(),
           'param': [{'var_smoothing': [10**exp for exp in range(-3,-12,-1)]}]
          },
    'lp': {'name': 'Linear Perceptron   ',
           'estimator': Perceptron(random_state=random_state),
           'param': [{'early_stopping': [True,False],'class_weight':[None,'balanced']}],
          },
    'svc':{'name': 'Support Vector      ',
           'estimator': SVC(random_state=random_state),
           'param': [{'kernel': ['rbf'],
                    'gamma': [1e-3, 1e-4],
                    'C': [1, 10, 100],
                    },
                    {'kernel': ['linear'],
                     'C': [1, 10, 100],
                    },
                   ]
          },
    'knn':{'name': 'K Nearest Neighbor ',
           'estimator': KNeighborsClassifier(),
           'param': [{'n_neighbors': list(range(1,7))}]
       },
    'adb':{'name': 'AdaBoost           ',
           'estimator': AdaBoostClassifier(random_state=random_state),
           'param': [{'n_estimators':[20,30,40,50]
                     ,'learning_rate':[0.5,0.75,1,1.25,1.5]}]
          },
    'rf': {'name': 'Random forest       ',
           'estimator': RandomForestClassifier(random_state=random_state),
           'param': [{
                      # 'max_depth': [*range(4,30)]
                      'max_depth': [*range(4,30,4)]
                    #  ,'n_estimators':[*range(20,80,5)]
                     ,'n_estimators':[*range(10,30,5)]
                     ,'class_weight': [None,'balanced']
                      }]
          }

}
```

```python
pca = PCA()
X_trans = pca.fit_transform(X_full)
variance = pca.explained_variance_ratio_
plt.plot(range(X_full.shape[1]),variance,color="pink")
plt.title("Variance Ratio")
plt.xlabel("Columns")
plt.ylabel("Variance")
plt.ylim(0,0.12)
plt.show()
```

```python
# Sort the explained variance ratios in descending order


# Calculate the cumulative sum of the explained variance ratios

# Find the index i of the first principal component where the cumulative sum exceeds 0.05

# Keep the first i principal components and discard the rest

variance = sorted(variance,reverse = True)
cum_sum = 0
i = 0
while cum_sum < 0.9:
    cum_sum += variance[i]
    i+=1
print(f"We need {i} components to have a cumulative sum of explained variance ratio of 0.9")
```

```python
cum_sum = np.cumsum(variance)
cutoff = 0.8
cutoff_index = np.argmax(cum_sum > cutoff)
X_cut = pca.transform(X_full)[:,:cutoff_index+1]
```

```python
pca = PCA(44)
X_trans = pca.fit_transform(X_full)
variance = pca.explained_variance_ratio_
plt.plot(variance,color="pink")
plt.title("Variance Ratio")
plt.xlabel("Columns")
plt.ylabel("Variance")
plt.ylim(0,0.12)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_cut, y, train_size = 2000)
```

```python
y.unique()
```

```python
classes_train,counts_train = np.unique(y_train, return_counts=True)
classes_test,counts_test = np.unique(y_test, return_counts=True)
classes_train=[i[-1] for i in classes_train]
classes_test=[i[-1] for i in classes_test]
fig, axs = plt.subplots(1,2)
axs[0].pie(counts_train, labels = classes_train)
axs[1].pie(counts_test, labels = classes_test)
plt.tight_layout()
plt.show()
```

```python
scorings = ['accuracy'
            # ,'precision_macro'
            # ,'recall_macro'
            ,'f1_macro'
            ]
```

```python
from sklearn.model_selection import ParameterGrid
trials = {lbl : len(ParameterGrid(models[lbl]["param"])) for lbl in model_lbls}
print(trials)
```

```python
best_params=[]
y_preds=[]
results = pd.DataFrame(columns=['scoring','model','best_params','accuracy','precision_macro','recall_macro','f1_macro'])

```

```python
#grid search cv
for scoring in scorings:
    for model in model_lbls:
        clf = GridSearchCV(models[model]["estimator"],models[model]["param"],scoring=scoring,cv=3)
        clf.fit(X_train,y_train)
        best_param = clf.best_params_
        best_score = clf.best_score_
        y_pred = clf.predict(X_test)
        cv = classification_report(y_test, y_pred, output_dict=True)
        results.loc[len(results)] = [scoring,models[model]["name"],best_param,cv["accuracy"],cv["macro avg"]["precision"],cv["macro avg"]["recall"]
                                 ,cv["macro avg"]["f1-score"]]
        best_params.append(best_param)
        y_preds.append(y_pred)
    print(f"JUST FOR DEBUGGING FINISHED :{scoring} !")

```

```python
print("Best results for accuracy")
result_acc = results.sort_values(by="accuracy",ascending=False).reset_index()
result_acc
```

```python
print("Best results for f1_macro")
result_f1 = results.sort_values(by="f1_macro",ascending=False).reset_index()
result_f1
```

```python
result_acc.loc[0,"best_params"]
```

```python
y_pred_acc = y_preds[best_params.index(result_acc.loc[0,"best_params"])]
disp = ConfusionMatrixDisplay(confusion_matrix(y_test,y_pred_acc,normalize="true"))
disp.plot()
plt.show()
```

```python
y_pred_f1 = y_preds[best_params.index(result_f1.loc[0,"best_params"])]
disp = ConfusionMatrixDisplay(confusion_matrix(y_test,y_pred_f1,normalize="true"))
disp.plot()
plt.show()
```
